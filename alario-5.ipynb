```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sidetable

```

```python
df = pd.read_csv("data/4_tabla_sin_nulos_sin_outliers.csv", index_col = 0)
df.head(2)
```

```python
df_re_filtrado = df[df["continente"] != "Otros"]
```

Hacemos el filtro para ver las entradas por continente

```python
proporcion_continente = df_re_filtrado.stb.freq(["continente"])
proporcion_continente
```

```python
plt.figure(figsize=(6,6))
plt.pie(proporcion_continente["count"], labels=proporcion_continente["continente"], autopct="%1.2f%%", radius=2)
plt.show()
```

Miramos de cada continente los paises donde se cobra más

```python
plt.figure( figsize=(20,10))
sns.set(style="darkgrid")
plt.title(label="Asia")

plt.xticks(rotation=70)


sns.barplot(data=df[df["continente"] == "Asia" ], y="rango_salario_maximo", x="pais", palette="rainbow_r", errorbar=None)
    
plt.grid(color = 'grey' , linestyle= '-.', linewidth = 0.3,)

plt.legend(loc = "best", bbox_to_anchor = (1,1))
```

```python
fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(15,30))
sns.set(style="darkgrid")
plt.title(label="Asia")
axes = axes.flat

lista_columnas = ['edad', 'genero','años_programando','sector_trabajo_actual', 'puesto_trabajo']
for i, col in enumerate(lista_columnas):
    sns.barplot(data=df[df["pais"] == "Israel"], y="rango_salario_maximo", x="pais", ax=axes[i], palette="rainbow_r", hue=col, ci=None)
    # ci nos quita la barrita de std por cada una de las columnas
    plt.grid(color = 'grey' , linestyle= '-.', linewidth = 0.3,)
    axes[i].set_title(col)
axes[0].legend(loc = "best", bbox_to_anchor = (1,1))
axes[1].legend(loc = "best", bbox_to_anchor = (1,1))
axes[2].legend(loc = "best", bbox_to_anchor = (1,1))
axes[3].legend(loc = "best", bbox_to_anchor = (1,1))
axes[4].legend(loc = "best", bbox_to_anchor = (1,1))
```
