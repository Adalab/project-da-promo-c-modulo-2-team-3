```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sidetable

plt.rcParams["figure.figsize"]=(30,15)
```

```python
df = pd.read_csv("data/4_tabla_sin_nulos_sin_outliers.csv", index_col = 0)
df.head(2)
```

```python
df_asia = df[df["continente"] == "Asia"]
df_asia.head(2)
```

```python
df_africa = df[df["continente"] == "Africa"]
df_africa.head(2)
```

```python
df_europa = df[df["continente"] == "Europa"]
df_europa.head(2)
```

```python
df_america = df[df["continente"] == "America"]
df_america.head(2)
```

```python
df["continente"].value_counts()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10,15))
axes = axes.flat
lista_continentes = ["Asia", "Africa", "Europa", "America", "Oceania"]
for i, col in enumerate(lista_continentes):
    #print(f"{i} --- {col}")
    sns.boxplot(y="pais", x="rango_salario_maximo", data=df[df["continente"] == col], ax=axes[i])
    plt.grid(color = 'grey' , linestyle= '-.', linewidth = 0.3)
    #axes[i].tick_params(rotation = 90)
    axes[i].set_title(col)

# ajusta el tamaño
#fig.tight_layout()
# eliminamos la última gráfica que está vacía
fig.delaxes(axes[-1])
```
